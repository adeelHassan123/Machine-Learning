```python
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,SGDRegressor
from sklearn.preprocessing import PolynomialFeatures,StandardScaler
from sklearn.metrics import r2_score
```

```python
X = 6 * np.random.rand(200, 1) - 3
y = 0.9 * X**2 + 0.3 * X + 2 + np.random.randn(200, 1)
```

```python
plt.plot(X, y,'b.')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train, y_train,'b.')
plt.plot(X_train,lr.predict(X_train),color='r')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.fit_transform(X_test)
```

```python
lr = LinearRegression()
lr.fit(X_train_scaled,y_train)
y_pred = lr.predict(X_test_scaled)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train_scaled, y_train,'b.')
plt.plot(X_train_scaled,lr.predict(X_train_scaled),color='r')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
poly = PolynomialFeatures(degree=2,include_bias=True)
X_train_poly = poly.fit_transform(X_train_scaled)
X_test_poly = poly.transform(X_test_scaled)
lr = LinearRegression()
lr.fit(X_train_poly,y_train)
y_pred = lr.predict(X_test_poly)
r2_score(y_test,y_pred)
```

```python
plt.plot(X_train_poly, y_train,'b.')
plt.plot(X_train_poly,lr.predict(X_train_poly),color='r')
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
print(lr.coef_)
print(lr.intercept_)
```

```python
X_new=np.linspace(-3, 3, 200).reshape(200, 1)
X_new_poly = poly.transform(X_new)
y_new = lr.predict(X_new_poly)
```

```python
plt.plot(X_new, y_new, "r-", linewidth=2, label="Predictions")
plt.plot(X_train, y_train, "b.",label='Training points')
plt.plot(X_test, y_test, "g.",label='Testing points')
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
