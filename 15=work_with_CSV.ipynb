<h1>Opening CSV from open URL
</h1>

```python
import requests
import pandas as pd
from io import StringIO

# URL of the CSV file
URL = "https://raw.githubusercontent.com/google/dspl/master/samples/google/canonical/countries.csv"

# Get the data from the URL
response = requests.get(URL)

# Convert the response text into a file-like object
data = StringIO(response.text)

# Read the CSV using pandas
df = pd.read_csv(data)

# Display the dataframe
df.head()
```

<h1>
Converters
</h1>

```python
#to change ling names to short form e.g PAKISTAN MUSLIM LEAGUE NOON =PMLN

def rename(name):
    if name == "PAKISTAN MUSLIM LEAGUE":
        return "pmln"
    else:
        return name


pd.read_csv("placement.csv",converters={"col_name": rename})
    
```
