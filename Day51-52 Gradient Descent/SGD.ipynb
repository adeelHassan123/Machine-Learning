```python
import numpy as np
from sklearn.datasets import load_diabetes 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import time
```

```python
def cost_error_function(X, y, w_in, b_in):
    cost = y - (np.dot(X,w_in) + b_in)
    return cost
```

```python
def gradient_Slope_function(X, y, w_in, b_in, cost_error_function):
    cost = cost_error_function(X, y, w_in, b_in)
    slope_b = -2 * np.mean(cost)
    slope_w = -2 * np.dot(X.T , cost) / len(y)
    return slope_b,slope_w
```

```python
class my_GD:
    def gradient_descent(self, X, y, w_in, b_in, cost_error_function, gradient_Slope_function, LearningRate_alpha, epochs):
        self.b_in = b_in
        self.w_in = w_in
        J_history =[]
        for i in range(epochs):  
            slope_b,slope_w = gradient_Slope_function(X, y, self.w_in, self.b_in, cost_error_function)
            self.b_in -= LearningRate_alpha * slope_b
            self.w_in -= LearningRate_alpha * slope_w  # ‚Üê CRITICAL UPDATE STEP
                    # Save cost J at each iteration
            if i<100000:      # prevent resource exhaustion 
                J_history.append( cost_error_function(X, y, self.w_in, self.b_in,))

        print("My Linear Regression Coefficients:", self.b_in)
        print("My Linear Regression Intercept:", self.w_in)
        return J_history

    def predict(self, X):
        X = np.array(X)
        print(self.b0 + X.dot(self.b))
```

```python
Xx,yy = load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(Xx,yy)

```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
print("Sklearn Linear Regression Coefficients:", lr.coef_)
print("Sklearn Linear Regression Intercept:", lr.intercept_)


# initialize parameters
initial_w = np.zeros(X_train.shape[1])
initial_b = 0.

my_GD = my_GD()
# run gradient descent 
J_history = my_GD.gradient_descent(X_train, y_train, initial_w, initial_b, cost_error_function, gradient_Slope_function, LearningRate_alpha=0.01, epochs=1000)
```

```python
# plot cost versus iteration  
fig, (ax1, ax2) = plt.subplots(1, 2, constrained_layout=True, figsize=(12, 4))
ax1.plot(J_history)
ax2.plot(100 + np.arange(len(J_history[100:])), J_history[100:])
ax1.set_title("Cost vs. iteration");  ax2.set_title("Cost vs. iteration (tail)")
ax1.set_ylabel('Cost')             ;  ax2.set_ylabel('Cost') 
ax1.set_xlabel('iteration step')   ;  ax2.set_xlabel('iteration step') 
plt.show()
```
